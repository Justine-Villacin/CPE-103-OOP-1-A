# Calculation of Grades

```python
# Calculating Grade

class Calculate():

  def __init__(self, name):

    self.name = name
    self.PG, self.MG, self.FG = 0, 0, 0
    self.PQ, self.MQ, self.FQ = 0, 0, 0
    self.PHOA, self.MHOA, self.FHOA = 0, 0, 0
    self.PA, self.MA, self.FA = 0, 0, 0
    self.PE, self.ME, self.FE = 0, 0, 0
    self.PCS, self.MCS, self.FCS = 0, 0, 0

    # calculation of prelim class standing
  def calc_PCS(self):
      return (0.5 * self.PHOA) + (0.3 * self.PQ) + (0.2 * self.PA)

    # calculation of midterm class standing
  def calc_MCS(self):
      return (0.5 * self.MHOA) + (0.3 * self.MQ) + (0.2 * self.MA)

    # calculation of final class standing
  def calc_FCS(self):
      return (0.5 * self.FHOA) + (0.3 * self.FQ) + (0.2 * self.FA)

    # Calculating the PrelimGrade, MidterGrade, FinalGrade, and General Average
  def calc_PrelimGrade(self):
      self.PrelimGrade = (0.50 * self.PE) + (0.50 * self.calc_PCS()) # Call calc_PCS
      return self.PrelimGrade

  def calc_MidtermGrade(self):
      self.MidtermGrade = ((1/3) * self.PrelimGrade) + ((2/3) * ((0.50 * self.ME) + (0.50 * self.calc_MCS()))) # Call calc_MCS
      return self.MidtermGrade

  def calc_FinalGrade(self):
      self.FinalGrade = ((1/3) * self.MidtermGrade) + ((2/3) * ((0.50 * self.FE) + (0.50 * self.calc_FCS())))
      self.GenAve =  ((self.PrelimGrade + self.MidtermGrade + self.FinalGrade ) /3 )
      return self.FinalGrade, self.GenAve

  def inputgrade(self, SemGrade):
      print(f"Enter grades for {SemGrade}")

      while True:
        try:
          quiz = float(input(f" {SemGrade} Quiz: "))
          HOA = float(input(f" {SemGrade} Hands-on Activity: "))
          assignment = float(input(f"{SemGrade} Assignment: "))
          exam = float(input(f"{SemGrade} Exam: "))

          # Checking the number Input
          if 0 <= quiz <= 100 and 0 <= HOA <= 100 and 0 <= assignment <= 100 and 0 <= exam <= 100:
            break
          else :
            print(f"Enter a valid grade out of 100")
        except ValueError:
          print(f"Enter a valid grade out of 100" )


      if SemGrade == "Prelim":
        self.PQ = quiz
        self.PHOA = HOA
        self.PA = assignment
        self.PE = exam
      elif SemGrade == "Midterm":
        self.MQ = quiz
        self.MHOA = HOA
        self.MA = assignment
        self.ME = exam
      elif SemGrade == "Final":
        self.FQ = quiz
        self.FHOA = HOA
        self.FA = assignment
        self.FE = exam
for x in range(3):
  # Calculating Grade
# Student Name
  StudentName = input(f"Enter your name: ")
  StudentNumber = input(f"Enter your student number: ")
  Course = input(f"Enter your course: ")
  result = Calculate(StudentName)

  # Input Semester Grades
  for SemGrade in ["Prelim", "Midterm", "Final"]:
    result.inputgrade(SemGrade)

  # Calculate Prelim
  PrelimGrade = result.calc_PrelimGrade()

  # Calculate Midterm
  MidtermGrade = result.calc_MidtermGrade()

  #Display the final Output
  FinalGrade, GenAve = result.calc_FinalGrade()

  print(f"\nYour prelim grade is: {PrelimGrade:.2f}")
  print(f"Your midterm grade is: {MidtermGrade:.2f}")
  print(f"Your final grade is: {FinalGrade:.2f}")
  print(f"\nYour general average is: {GenAve:.2f}")

  # Convert Grade Value to UCC
  def grade_conversion_description(grade):
    if grade >= 99:
      return "Excellent - Grade Equivalent is: 1.00"
    elif grade >= 90:
      return "Very Good - Grade Equivalent is: 1.25 - 1.50"
    elif grade >= 80:
      return "Good - Grade Equivalent is: 1.75 - 2.00"
    elif grade >= 70:
      return "Satisfactory - Grade Equivalent is: 2.25 - 2.50"
    elif grade >= 60:
      return "Passing - Grade Equivalent is: 2.75 - 3.00"
    else:
      return "Failed - Grade Equivalent is: 5.00"

  description = grade_conversion_description(GenAve)
  print(f"Your grade description is:", description)
  print("\n")
```
