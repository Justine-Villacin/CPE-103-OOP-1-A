```python
# STUDENT INFORMATION

def main():

  studentID = input("Student ID: ")
  name = input("Name: ")
  program = input("Program: ")
  course = input("Course: ")
  sex = input("Sex: ")
  weight = float(input("Weight(kg): "))
  height = float(input("Height(cm): "))


  print(" ")
  # COMPUTATION for MIDTERM
  print("MIDTERM GRADE: ")
  Quiz = float(input("Quiz: "))
  Seatwork = float(input("Seatwork: "))
  Assignment = float(input("Assignment: "))
  Recitation = float(input("Recitation: "))

  MidtermGrade = (0.40 * Quiz) + (0.30 * Seatwork) + (0.20 * Assignment) + (0.10 * Recitation)


  print(" ")
  print(f"Midterm Grade: {MidtermGrade:.2f}")

  # COMPUTATION for FINAL GRADE
  print(" ")
  print("FINAL GRADE: ")
  Quiz = float(input("Quiz: "))
  Seatwork = float(input("Seatwork: "))
  Assignment = float(input("Assignment: "))
  Recitation = float(input("Recitation: "))

  print(" ")
  FinalGrade = (0.40 * Quiz) + (0.30 * Seatwork) + (0.20 * Assignment) + (0.10 * Recitation)


  print(f"Final Grade: {FinalGrade:.2f}")

  # CALCULATING GWA
  print(" ")

  print("---------------------------------------------------\n")
  gwa = (MidtermGrade + FinalGrade)/2
  print(f"GWA: {gwa:.2f}")
  print(f"Course Enrolled: ", course)
  print("\n---------------------------------------------------")

  print(" ")
  while True:
      output = input(f"Another Student? Yes or No: ")
      print(" ")
      if output == "Yes":
        main()
        break
      elif output == "No":
        print(" ")
        print("Thank You")
        return
      else:
        print(" ")
        print("Invalid Input")

main()
```
